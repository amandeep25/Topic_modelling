```python
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
from top2vec import Top2Vec
from gensim.models import CoherenceModel
from gensim import corpora
from sklearn.feature_extraction.text import CountVectorizer

# Load 20 Newsgroups dataset
newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))

# Create a DataFrame
df = pd.DataFrame({'text': newsgroups.data, 'category': newsgroups.target})

# Optional: Preprocess the text data if needed

# Extract the text as a list of strings
documents = df['text'].tolist()

# Create and fit Top2Vec model
model = Top2Vec(documents, embedding_model='universal-sentence-encoder')

# Get topics and keywords
topics, _, _ = model.get_topics()

# Display topics and their keywords
for i, topic in enumerate(topics):
    print(f"Topic {i + 1}: {', '.join(topic)}")

#
```

https://github.com/ddangelov/Top2Vec/issues/158

```python
documents_df = pd.DataFrame({"ProcessedTexts": df['text'], "ID": df.index, "Topic": model.doc_top})
#change the column name df text to be 'ProcessedTexts'
```

```python
def t2vCoherence(df, topic_words):
   tokenized = [list(tokenize(doc)) for doc in df.ProcessedTexts.tolist()]
   id2word = corpora.Dictionary(tokenized)
   corpus = [id2word.doc2bow(text) for text in tokenized]
   # make sure you grab the topic words from the topic model and convert them to a list
   coherence_model = CoherenceModel(topics= topic_words ,texts=tokenized
               , corpus=corpus, dictionary=id2word, coherence='c_v', topn=50)  #Use top 50 words since Top2Vec gets top 50 words
   coherence = coherence_model.get_coherence()
   print("Model Coherence C_V is:{0}".format(coherence))
   return coherence
```

```python
def t2vCoherence_umass(df, topic_words):
   tokenized = [list(tokenize(doc)) for doc in df.ProcessedTexts.tolist()]
   id2word = corpora.Dictionary(tokenized)
   corpus = [id2word.doc2bow(text) for text in tokenized]
   # make sure you grab the topic words from the topic model and convert them to a list
   coherence_model = CoherenceModel(topics= topic_words ,texts=tokenized
               , corpus=corpus, dictionary=id2word, coherence='u_mass', topn=50)  #Use top 50 words since Top2Vec gets top 50 words
   coherence = coherence_model.get_coherence()
   print("Model Coherence C_V is:{0}".format(coherence))
   return coherence
```
